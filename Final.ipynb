Find Maximum depth and width of trapezoidal open channel 


<h3> Manning's Equation </h3>

$$ Q = \frac{1.49}{n}AR^{2/3}\sqrt{S} $$

```python
A = b*y + z*y^2 # Area for trapezoidal channel
```

```python
P = b + (2*y*(1+z^2))^2 # perimeter for trapezoidal channel
```

```python
R = A/P
```

```python
Q = (1.49/n)*A*(R^0.67)*S^0.5 # manning equation
```
